```python
df=data.copy()
df.info()
```

```python
#transform non numeric label into numeric label
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=l.fit_transform(df[i])


```

```python
df.shape
```

```python
#splitting dataset
X=df.drop('Mental_fitness',axis=1)
y=df["Mental_fitness"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=.20,random_state=2)
print("xtrain",xtrain.shape)
print("xtest",xtest.shape)
print("\n ytrain",ytrain.shape)
print("ytest",ytest.shape)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
#fitting training data into model
lr.fit(xtrain,ytrain)
ytrain_pred=lr.predict(xtrain)
mse=mean_squared_error(ytrain,ytrain_pred)
rmse=(np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2=r2_score(ytrain,ytrain_pred)
print("The linear regression model performance of training set")
print("Mean Squared Error is {}".format(mse))
print("Root Mean squared Error is {}".format(rmse))
print("R2 score is {}".format(r2))
```
